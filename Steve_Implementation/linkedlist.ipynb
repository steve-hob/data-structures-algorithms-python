Linked list are series of nodes that makes it easier to insert or remove elements from any where in the middle of the list

```python
class Node:
    # each node in a linked list has a value and the index to the next value
    def __init__(self, val = None, idx = None):
        self.value = val
        self.next = idx
```

```python
class LinkedList:

    ### this implmentation makes LL look kinda like this:
    ### [1,[2,[3,[4,[5,None]]]]]

    def __init__(self):
        self.head = None

    def print_ll(self):
        node = self.head
        ll = ''
        
        while node.next:
            ll += str(node.value) + '--'
            node = node.next

        ll += str(node.value)
        print(ll)
    
    def get_len(self):
        count = 0
        node = self.head

        while node:
            node = node.next
            count += 1
        
        print('L.L. Length =',count)
    
    def insert_at_bgn(self, data):
        self.head = Node(data, self.head)

    def insert_at_end(self, data):
        node = self.head 

        while node.next:
            node = node.next 
        node.next = Node(data, None)

    def insert_values(self, node_list):
        for elem in node_list:
            self.insert_at_end(elem)

    def insert_at(self, data, idx):
        node = self.head
        count = 0

        if idx == 0:
            self.insert_at_bgn(data)

        else:
            # stopping at the previous node
            # does not handle 0
            while count != idx-1: 
                node = node.next
                count += 1

            new_node = Node(data, node.next)
            node.next = new_node

    def remove_at(self, idx):
        node = self.head
        
        if idx == 0:
            self.head = self.head.next

        else:
            count = 0
            while count <= idx-1: # takes me to the node immediately before the removal node
                node = node.next
                count += 1
            node.next = node.next.next

    def insert_after_value(self, val, data):
        node = self.head

        idx = 0
        while node.value != val:
            node = node.next
            idx += 1
            # lacking an end node handle
        self.insert_at(data, idx+1)


    def remove_by_value(self, val):  
        node = self.head

        idx = 0
        while node.value != val:
            node = node.next
            idx += 1
        self.remove_at(idx-1)

if __name__ == '__main__':
    ll = LinkedList()

    ll.insert_at_bgn(2)
    ll.insert_at_bgn(1)
    ll.insert_values([4,4,5])
    ll.print_ll()

    ll.insert_at(3,2)
    ll.insert_at(0,0)
    ll.print_ll()

    ll.remove_at(4)
    ll.remove_at(0)
    ll.print_ll()

    ll.insert_after_value(5, 6)
    ll.print_ll()
    ll.remove_by_value(6)
    ll.remove_by_value(3)
    ll.print_ll()

    print('='*10)
    ll.get_len()
    ll.print_ll()
```
